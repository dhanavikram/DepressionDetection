```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Feature Selection for Depression Detection

```python
train_df = pd.read_csv("train_data_final.csv")
train_df.info()
```

## Creating a correlation matrix

```python
corr_matrix = train_df.corr()

plt.figure(figsize=(10, 8))

sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm',
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# find all correlations that are above magnitude 0.5

fltr_matrix = corr_matrix.abs() >= 0.5
correlated_cols = corr_matrix.columns[fltr_matrix.any()]

fltr_matrix = corr_matrix.loc[correlated_cols, correlated_cols]
```

```python
fltr_matrix = corr_matrix.abs() >= 0.5
# print(fltr_matrix)

# correlated_cols = corr_matrix.columns[fltr_matrix]

new_matrix = corr_matrix[fltr_matrix]
print(len(new_matrix))
```
